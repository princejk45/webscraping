```python
!pip install xlsxwriter

from bs4 import BeautifulSoup
import requests
import xlsxwriter
```

```python
page_to_scrape = requests.get("https://www.scrapethissite.com/pages/simple/")
soup = BeautifulSoup(page_to_scrape.text, "html.parser")

countries = soup.findAll("h3", attrs={"class": "country-name"})
capitals = soup.findAll("span", attrs={"class": "country-capital"})
populations = soup.findAll("span", attrs={"class": "country-population"})
areas = soup.findAll("span", attrs={"class": "country-area"})

# Create a new workbook
workbook = xlsxwriter.Workbook("country_scraped.xlsx")
worksheet = workbook.add_worksheet()

row = 0
col = 0

for country, capital, population, area in zip(countries, capitals, populations, areas):

    # Print Output
    print(country.text.strip() + " - " + capital.text + " - " + population.text + " - " + area.text)

    # Write data to each cell in the row

    worksheet.write(row, col, country.text.strip())
    worksheet.write(row, col + 1, capital.text)
    worksheet.write(row, col + 2, population.text)
    worksheet.write(row, col + 3, area.text)


    # Move to the next row
    row += 1

# Close the workbook to save the changes
workbook.close()
```
